```python
### 1. What types are involved in the following expressions? What data type will each expression evaluate to?

Output = 8 * 2.5  # --> multiplication
print('Output: ', Output, '\n', 'Data type: ', type(8),type(2.5),type(Output))
```

```python
Output = 9 / 2 # --> division
print('Output: ', Output, '\n', 'Data type: ', type(9),type(2),type(Output))
```

```python
Output = 9 // -2 # --> floor division, rounding to the nearest whole number (usually rounding up)
print('Output: ', Output, '\n', 'Data type: ', type(9),type(-2),type(Output))
```

```python
Output = (1.5 * 2 >= 7 - 3) # --> False, because 3 is not larger than 4
print('Output: ', Output, '\n', 'Data type: ', type(1.5),type(2), type(7),type(3),type(Output))
```
