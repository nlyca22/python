### 2. Append a "world" to the following word.

```python
word = 'Hello'
word = word.append(' World')
print(word)

# The above code results in error, because .append() method is used for lists, not strings.

```

```python
word = 'Hello'
word += ' World'
print(word)  # This will print: Hello World

# The above code appends string to another string
```
