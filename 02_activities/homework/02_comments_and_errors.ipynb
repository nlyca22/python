````python
84 *= 0.5 / 7 

# The *= operator in Python is an assignment operator that performs multiplication and assignment in one step. It multiplies the variable by a specified value and then assigns the result back to that variable.

# For the expression 84 *= 0.5 / 7:

# First, the division 0.5 / 7 is calculated, which equals approximately 0.07142857.
# Then, 84 is multiplied by 0.07142857, resulting in approximately 6.0.
# Finally, 84 is updated to 6.0 --> which will result in an error, because it's illegal to force 84 to be 6. 
````

<details>
  <summary>Answer</summary>

  Error -- Python thinks we are trying to assign a value to 84, which is an integer literal. We can't change its value!
