### 1. Write a conditional that prints different messages if a bank account balance is below `$`3,000, between `$`3,000 and `$`10,000, or over `$`10,000.

HINT: You can check if a value `x` is between two other values with a condition like `5 <= x <= 10`, but you can also solve this problem by strategically ordering conditions.

````python
# Your code goes here

balance = 2900

if balance < 3000:
    print('Balance is below $3,000')

elif 3000 <= balance <= 10000:
    print('Balance is within range $3,000 - $10,000')

else: 
    print('Balance is over $10,000')
````

### 3. Given the list `people`, sort it by people's first name, last name and age. Store the sorted lists as `by_first_name`, `by_last_name`, and `by_age`, respectively.  
`people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]`

HINT 1: We want our sorting function to return a new list, rather than modifying in place.

HINT 2: `sorted()` accepts a `key` argument. This argument is the name of a function to use when sorting list elements.

````python
# Your code goes here

people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)] # a list of tuples (not a 3X3 array)

sorted(people) # this automatically sorts by the first element it encounters, which is the 1st element in the tuples. "E" then "K" then "M".
````

````python
# We want to extract the 1st, 2nd, and 3rd elements within the tuples
# Fortunately, the DEF function can extract the elements within the tuples within the people list
# Coupled with the KEY feature within sorted(), the KEY looks at what were defined in the DEF functions 

people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]

# Creating 3 separate DEFINE functions: 
def first_element_within_tup(tup):
    return tup[0]

def second_element_within_tup(tup):
    return tup[1]

def third_element_within_tup(tup):
    return tup[2]

# Sort 
by_first_name = sorted(people, key=first_element_within_tup)
by_last_name = sorted(people, key=second_element_within_tup)
by_age = sorted(people, key=third_element_within_tup)


print("Sort by Firt Name: ", by_first_name)
print("Sort by Last Name: ", by_last_name)
print("Sort by Age: ", by_age)
````

````python
# A more elegant coding would be to use "lambda" in sorted() 

''' 
KEY FEATURES OF SORTED():

- Syntax: sorted(iterable, key=None, reverse=False)

- Parameters:
     iterable: The sequence (list, tuple, string, etc.) to be sorted.
     key (optional): A function that serves as a key for the sort comparison. For example, key=str.lower will sort strings in a case-insensitive manner.
     reverse (optional): A boolean value. If True, the sorted list is reversed (or sorted in descending order).


- Example: 
    people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]

    sorted_by_age = sorted(people, key=lambda element_within_tuple: element_within_tuple[2])
    
    print("Sort by Age: ", sorted_by_age)
    >>> Sort by Age: [('Ken', 'Wolseley', 23), ('Mark', 'Harrison', 56), ('Emily', 'Robinson', 77)]

- NOTE: 
     * sorted() returns a new sorted list and does not modify the original iterable.
     * list.sort() is a method that modifies the list in place and returns None.

'''
````

````python
people = [('Mark', 'Harrison', 56), ('Ken', 'Wolseley', 23), ('Emily', 'Robinson', 77)]

sorted_by_last_name = sorted(people, key=lambda element_within_tuple: element_within_tuple[1])
    
print("Sort by Last Name: ", sorted_by_last_name)
````

HINT: Let's break this down into steps. We need to:

*   Convert them both to sets
*   Find their intersection

HINT: Some useful functions and methods are `keys()`, `set()`, and `intersection()`.

<details>
  <summary>Answer</summary>

  ```python
  def dict_intersect(d1, d2):
    '''Return the set of keys found in both d1 and d2.
    >>> dict_intersect({'a': 'A', 'b': 'B', 'c': 'C'}, {'a': 'alpha', 'b': 'beta'})
    {'a', 'b'}
    >>> dict_intersect({'a': 1, 'b': 2}, {'c': 3, 'd': 2})
    set()
    '''
    d1_keys = set(d1.keys())
    d2_keys = set(d2.keys())
    return d1_keys.intersection(d2_keys)
  ```

